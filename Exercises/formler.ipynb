```python
import numpy as np
from scipy.stats import binom
import math
import matplotlib.pyplot as plt

def binom_1(n, p, k):
    prob = math.comb(n, k)*p**k*(1-p)**(n-k)
    return round(prob, 4)

def binom_2(n, p, k):
    prob = binom.pmf(k, n, p)
    return round(prob, 4)


def factorial(n):
    nfact = 1
    for x in range(n+1):
        if x != 0:
            nfact *= x

    return nfact

def permut(n, r):
    return factorial(n)/factorial(n-r)

def comb(n, r):
    return factorial(n)/(factorial(n-r)*factorial(r))

def mean(values):
    return round(sum(values)/len(values), 6)

def std(values):
    sqDiffsum = 0
    m = mean(values)
    n = len(values)
    for x in values:
        sqDiffsum += (x - m)**2
    
    std = np.sqrt(sqDiffsum/(n-1))

    return round(std, 6)


```

```python

matchprob = []

for n in range(100):
    noMatch = 1
    for x in range(n):
        noMatch *=(365 - x)/365
    matchprob.append(1-noMatch)


plt.figure(figsize=(6,4))
plt.plot(list(range(100)), matchprob, color="#ff0000")
plt.title("The birthday problem")
plt.xlabel("Number of people in the room")
plt.ylabel("probability of at least one shared birthday")
plt.tight_layout()
```

```python
n=50
p=0.5

x_values = list(range(n + 1))
y_values = [binom_2(n, p, k) for k in x_values]

coinflips = []

for i in range(1000):
    heads = np.random.randint(0,2,size=n).sum()
    coinflips.append(heads)


fig, ax = plt.subplots(1,2, figsize=(9,3))
ax[0].bar(x_values, y_values)
ax[1].hist(coinflips, bins=n, density=True)
plt.show()
```
